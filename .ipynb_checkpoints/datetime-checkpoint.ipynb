# Datetime

```python
import datetime
# datetime.date will have year, month, second
d = datetime.date(2016,7,24) # year,month,date (write plain integer , if we give 07 instead of 7 this will give error)
print(d)

today = datetime.date.today()# to get today's local date
print(today)
print(today.year) # to get year of today's
print(today.day) # to get day of today's
print(today.weekday()) # to get week of the day (monday is 0 and sunday is 6)
print(today.isoweekday()) # to get week of the day(monday is 1 and sunday is 7)
```

```python
# timedelta

tdelta = datetime.timedelta(days=7) # gives interval of time or date

print(today + tdelta)
print(today - tdelta)

# date2 = date1 + timedelta
# timedelta = date1 + date2

bday = datetime.date(2025, 1 ,14)

till_bday = bday - today
print(till_bday,type(till_bday))
print(till_bday.days)
print(till_bday.total_seconds())
```

```python
# datetime.time will have hour, minutes, seconds, microseconds

t= datetime.time(9, 30, 45, 100000)
print(t)
print(t.hour)
print(t.minute)
print(t.second)
print(t.microsecond)

dt = datetime.datetime(2025, 1, 14, 12, 30, 45, 100000) # gives both date and time
print(dt)
print(dt.date())
print(dt.time())
print(dt.year)
print(dt + tdelta)
tdelta = datetime.timedelta(hours=12)
print(dt + tdelta)

dt_today = datetime.datetime.today() # gives current local datetime without timezone
dt_now = datetime.datetime.now() # gives current local datetime with timezone, TZ info is set to None by default
dt_utcnow = datetime.datetime.utcnow() # UTC standard time, TZ info is set to None by default

print(dt_today)
print(dt_now)
print(dt_utcnow)
```

```python
import pytz # we should work with UTC date when dealing with timezone

dt = datetime.datetime(2016, 7, 27, 12, 30,45, tzinfo = pytz.UTC)# for given date and time convert to UTC timezone
print(dt)
dt_now = datetime.datetime.now(tz = pytz.UTC) # utc time with timezone
print(dt_now)
dt_utcnow = datetime.datetime.utcnow().replace(tzinfo = pytz.UTC) # utc time with timezone
print(dt_utcnow)

# to get india datetime timezone

dt_india = dt_now.astimezone(pytz.timezone('Asia/Kolkata')) # ist time with timezone
print(dt_india)

# to get list of all timezones present in pytz

for tz in pytz.all_timezones:
    print(tz)
```

```python
dt_india = datetime.datetime.now() # this date is the ist time but without timezone
print(dt_india)

# dt_tz = dt_india.astimezone(pytz.timezone('Asia/Kathmandu')) (this gives error because dt_india doesn't have timezone)

india_tz = pytz.timezone('Asia/kolkata')# grab your local timezone
print(india_tz)

dt_india = india_tz.localize(dt_india) # this date is IST time with timezone
print(dt_india)

dt_kat = dt_india.astimezone(pytz.timezone('Asia/Kathmandu')) # kathmandu time with timezone
print(dt_kat)

# to change date from one timezone to another 1st the input date must have timezone else it gives error
```

```python
print(dt_india.isoformat()) # this is one of the standard format followed worldwide

print(dt_india.strftime('%B %d, %Y')) # using strftime we can do date formatting, returns string from datetime refer docs

dt_str =  'February 02, 2024'

dt = datetime.datetime.strptime(dt_str, '%B %d, %Y') # to convert string to datetime
print(dt)

# strftime => Datetime to string
# strptime => string to datetime(need to also provide the format of date in string)

# pyarrow is another popular package to work with dates
```
